# What is NLP

- Natural Language Processing(NLP) is a field of Artificial Intelligence that gives machines the ability to read, understand and derive meaning from human language.
- With the help of NLP, we can communicate with computers using a natural language such as English.

# Steps to Solve NLP Problems

- <b>Gather Data</b>
    
   - Gather textual data from emails, posts or tweets.


- <b>Clean Data</b>
    
    - A clean dataset allows the model to learn meaningful features and not overfit on irrelevant noise.
    - Remove all irrelevant characters.
    - Tokenize the word by separating it in different words.


- <b>Clean Data</b>
    - Convert all characters to lowercase. 
    - Combine misspelled or alternatively spelled words to single representation.
    - Reduce words such as “am”, “are” and “is” to a common form.


- <b>Finding good representation</b>
    - Change textual data into numbers form which algorithms can understand and derive insights from it.


- <b>Classification</b>
    - Split the data into training and testing data.
    - Classify the data using a model by fitting the training data into it and check how well the model generalizes on unseen dataset using the testing dataset.


- <b>Inspection</b>
    - Understand the errors made by model using confusion matrix

## Polarity and Subjectivety

#### Polarity : 

- Expression which determines the sentimental aspect of an opnion. 
- Sentiment Polarity can be determined as positive, negative and neutral

- -1 ---> Negative Sentiment

```python
from textblob import TextBlob
```

```python
## let's calculate the polarity of the reviews

def get_polarity(text):
    textblob = TextBlob(str(text.encode('utf-8')))
    pol = textblob.sentiment.polarity
    return pol


## Let's apply the function
data['polarity'] = data['verified_reviews'].apply(get_polarity)
```

```python
## Let's create a dictionary for parts of Speech

pos_dic = {
    'noun' : ['NN','NNS','NNP','NNPS'],
    'pron' : ['PRP','PRP$','WP','WP$'],
    'verb' : ['VB','VBD','VBG','VBN','VBP','VBZ'],
    'adj' :  ['JJ','JJR','JJS'],
    'adv' : ['RB','RBR','RBS','WRB']
}
```

- NN  --> Singular Nouns
- NNS --> Plural Noun
- NNP --> Singular proper noun

```python
# ## Function to check and get tje part of speech tag count of words in a given sentence

# def pos_check(x, flag):
#     cnt = 0
#     try:
#         wiki = TextBlob(x)
#         for tup in wiki.tags:
#             ppo = list(tup)[1]
#             if ppo in pos_dic[flag]:
#                 cnt+=1
                
#     except:
#         pass
#     return cnt
```

```python
# function to check and get the part of speech tag count of a words in a given sentence
def pos_check(x, flag):
    cnt = 0
    try:
        wiki = TextBlob(x)
        for tup in wiki.tags:
            ppo = list(tup)[1]
            if ppo in pos_dic[flag]:
                cnt += 1
    except:
        pass
    return cnt
```

### Punctuation

```python
# Function to remove punctuation
def punctuation_removal(messy_str):
    clean_list = [char for char in messy_str if char not in string.punctuation]
    clean_str = ''.join(clean_list)
    return clean_str
```

### Numbers

```python
# Function to remove numbers for the reviews

## Import RegEx library
import re


def drop_numbers(list_text):
    ## Initiate an empty list
    list_text_new = []
    for i in list_text:
        ## Search empty list
        if not re.search('\d', i):
            list_text_new.append(i)
    return ''.join(list_text_new)         
```

```python
data['verified_reviews'] = data['verified_reviews'].apply(drop_numbers)
```

```python
##Function to remove the accented characters
## 1. Normalize the text to NFKD format
## 2. Change the text to ascii format, ascii does not have any accented characters

def remove_accented_chars(text):
    new_text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8','ignore')
    return new_text

##Let's apply the function
data2['verified_reviews'] = data2.apply(lambda x: remove_accented_chars(x['verified_reviews']), axis = 1)
```

## Stemming and Lemmitization

### Stemming

Stemming is a process of removing a part of a word, or reducing a word to its stem or root word.
Example:- 
- We have three words: “Ask”, “asking” and “asked”
- Stemming converts all the three words into the root word “ask”.


### Lemmatization

Lemmatization reduces the word to its dictionary form. The root word in lemmatization is called “lemma”.
- We have two words: “good” and “better”.
- Lemmatization reduce them to the same root word “good”.

## Difference between Stemming & Lemmatization

- Algorithms used in stemming process don’t know the meaning behind the words.
- Algorithms used in lemmatization process refer to a dictionary to understand the meaning of the word before chopping it off.

## Difference of Example

We have three words:- “play”, “playing” and “player”.
- According to Stemmer, all three words have same root word “play”.
- According to lemmatizer, “play” and “playing” have same root and “player” is a word with different meaning.


```python
##For Stemming
# import nltk

##For Lemitization
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
## Function for Stemming

def get_stem(text):
    stemmer = nltk.porter.PorterStemmer()
    text = ' '.join([stemmer.stem(word) for word in text.split()])
    return text

## call function
get_stem('we are eating and swimming ; we have been eating and swimming ; he eats and swims ; he ate and swam')
```

Feature extraction means to extract and produce feature representations that are appropriate for the NLP task.

Features that can be extracted from the text are:-

Major feature extraction techniques for NLP are:-
- Bag of words representation

### Bag of Words

Way of extracting features from text for use in modelling

Bag of words approach is simple and flexible

- checks the <b>frequency of distinct words</b> occuring in the text
- All the <b>distinct words make columns of matrix</b> and values 0 or 1 indicate the presence of the word in the text

### TF-IDF

<b>Term Frequency-Inverse Document Frequency</b>

Designed to reflect how important a word is to a document in a collection for corpus

The TF-IDF value increases proportionally to the number of times word appears in a document.

##### TF-IDF Calculation

The value of TF-IDF is calculated by multiplying two metrics: 

- How many times a word appears in a document.
- Inverse document frequency of the word across a set of documents.
    - Inverse Document frequency

    - How common a word is in entire document set
        
            log(1+Total no. of documents/(1+no. of documents containing word))+1
            
<b>Higher TF-IDF score, more relevant the word is in that particular document</b>

##### Areas of Application

- Information Retrival

    - TF-IDF was invented for document search and is used to deliver results that are most relevant to what we are searching for.
- Keyword extraction

    - TF-IDF is also useful for keyword extraction. The highest scoring words for a document are the most relevant keywords.

### N-Gram Analysis

Sequence of n-words
N-grams is a sequence of n-words.

For Example: I love reading books.
- 1-gram or unigram will be:- “I”, “love”, “reading”, “books”.
- 2-gram or bigram will be:- “I love”, “love reading”, “reading books”.
- 3-gram or trigram will be:- “I love reading”, “love reading books”.

### Why N-Gram Analysis ?

N-grams of texts are extensively used in text mining and NLP tasks such as <b>Auto completion</b> of sentences and <b>Auto spell check</b>.
Example:- 
- Using a 3-gram analysis, a bot will understand the difference between “What’s the temperature” and “Set the temperature” which is not possible using 1-gram or 2-grams.


#### What is the Best Value of N for performing N Gram Analysis?

ANs --> 2

The process of analyzing the natural language text and then labelling the text with a predefined set of labels or tags

### Conditional Probability

Probability of an event occurring based on the occurrence of the previous event.

Example:

We have a bag of 5 balls. 2 Reb Balls and 3 Blue balls

Initially

    Probability of getting a Red ball
    - P(R) = 2/5

    Probability of getting a Blue ball
    - P(B) = 3/5

Now we remove a blue ball

    Probability of getting a Red ball
    - P(R) = 2/4

    Probability of getting a Blue ball
    - P(B) = 2/4



### Bayes Theorm

- P(A/B) = P(A and B) / P(B) => P(A and B) = P(A/B) P(B) 
- P(B/A) = P(B and A) / P(A) => P(B and A) = P(B/A) P(A) 

Equating both the terms, we get Bayes Theorem

- P(A/B) = [P(B/A) P(A)] / P(B)

### Importing Data

```python
#Read the tb delimited file
data_nb = pd.read_csv('C:\\Users\\asus\\Downloads\\Udemy\\Machine Learning\\Natural Language Processing\\Feature+Engineering\\2. Feature Engineering\\amazon_alexa.tsv', delimiter = '\t')

#Check the shape of the dataset
data_nb.shape
```

### Cleaning the document

```python
# cleaning the texts
# importing the libraries for Natural Language Processing

import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
## Cleaning the Data

corpus = []

for i in range(0, 3150):
    review = re.sub('[^a-zA-Z]', ' ', data_nb['verified_reviews'][i])
    review = review.lower()
    review = review.split()
    ps = PorterStemmer()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

### Feature Extraction

```python
# Extracting the Features using the Tfidf Vectorizer

# from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(max_features = 2500)
x = tf.fit_transform(corpus).toarray()
y = data_nb.iloc[:, 4].values

print(x.shape)
print(y.shape)
```

### Test-Train Split

```python
### splitting the data into training and testing sets

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 15)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

### Modelling

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print("Training Accuracy :", model.score(x_train, y_train))
print("Testing Accuracy :", model.score(x_test, y_test))
```
